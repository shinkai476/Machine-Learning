## **1. BOOTSRAP TUNGGAL**

```python
# [cite_start]1. Mengambil sampel bootstrap tunggal [cite: 48, 49]
petal_length = df['petal length (cm)'].values
np.random.seed(42)
# [cite_start]Ukuran sampel = jumlah data asli, dengan pengembalian [cite: 50, 51]
boot_tunggal = np.random.choice(petal_length, size=len(petal_length), replace=True)

# [cite_start]2. Menampilkan perbandingan mean dan median dalam bentuk tabel [cite: 52, 53, 54, 55]
display(pd.DataFrame({
    'Statistik': ['Mean', 'Median'],
    'Data Asli': [np.mean(petal_length), np.median(petal_length)],
    'Bootstrap Tunggal': [np.mean(boot_tunggal), np.median(boot_tunggal)]
}))
```

## **2. BOOSTARP BERULANG**

```python
# Fungsi untuk melakukan bootstrap berulang
def bootstrap_means(data, iterations):
    means = []
    for _ in range(iterations):
        sample = np.random.choice(data, size=len(data), replace=True)
        means.append(np.mean(sample)) # Hitung mean dari setiap sampel
    return np.array(means)

# Lakukan bootstrap: 100, 1000, 10000 kali
np.random.seed(42)
boot_100 = bootstrap_means(petal_length, 100)
boot_1000 = bootstrap_means(petal_length, 1000)
boot_10000 = bootstrap_means(petal_length, 10000)

# Menampilkan output dalam bentuk tabel perbandingan
tabel_bootstrap = pd.DataFrame({
    'Jumlah Bootstrap': [len(boot_100), len(boot_1000), len(boot_10000)],
    'Rata-rata Mean Bootstrap': [boot_100.mean(), boot_1000.mean(), boot_10000.mean()]
})

print("--- Hasil Bootstrap Berulang ---")
display(tabel_bootstrap)
```

## **3. Visualisasi Hasil Bootsrap**

```python
# Visualisasi Histogram Distribusi Mean Bootstrap
plt.figure(figsize=(10, 6))

sns.histplot(boot_100, bins=30, kde=True, color='skyblue', label='100 Iterasi', alpha=0.5)
sns.histplot(boot_1000, bins=30, kde=True, color='salmon', label='1000 Iterasi', alpha=0.4)
sns.histplot(boot_10000, bins=30, kde=True, color='lightgreen', label='10000 Iterasi', alpha=0.3)

# Tambahkan garis untuk mean asli
mean_asli = np.mean(petal_length)
plt.axvline(mean_asli, color='red', linestyle='dashed', linewidth=2, label=f'Mean Asli: {mean_asli:.2f}')

plt.title('Distribusi Mean Bootstrap (Perbandingan Iterasi)', fontsize=14)
plt.xlabel('Nilai Mean', fontsize=12)
plt.ylabel('Frekuensi', fontsize=12)
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
# Creating boxplots for comparison
sns.boxplot(data=[boot_100, boot_1000, boot_10000])
plt.xticks([0, 1, 2], ['100 Iterasi', '1000 Iterasi', '10000 Iterasi'])

# Adding original mean line
plt.axhline(np.mean(petal_length), color='red', linestyle='--', label=f'Mean Asli: {np.mean(petal_length):.2f}')

plt.title("Boxplot Distribusi Mean Bootstrap", fontsize=14)
plt.ylabel("Nilai Mean")
plt.legend()
plt.show()
```

Berikut adalah perbandingan hasil Bootstrap untuk 100, 1.000, dan 10.000 iterasi:

### Bootstrap 100 Iterasi

Histogram (Garis Biru): Karena frekuensinya sangat rendah dibandingkan yang lain, distribusinya nyaris terlihat datar pada skala grafik ini. Jika diperbesar, bentuk kurvanya biasanya masih kasar dan belum membentuk lonceng sempurna.

Boxplot (Kotak Biru): Memiliki rentang nilai ekstrem (panjang whisker) yang paling pendek. Jumlah outlier (titik di luar batas) sangat sedikit (hanya terlihat satu atau dua). Hal ini terjadi karena 100 sampel acak belum cukup banyak untuk menangkap probabilitas kombinasi sampel yang ekstrem.

### Bootstrap 1.000 Iterasi

Histogram (Kurva/Area Merah Muda): Bentuk distribusi normal (kurva lonceng) sudah mulai terlihat jelas dan lebih simetris dibandingkan 100 iterasi. Frekuensinya memuncak di sekitar angka 60-80.

Boxplot (Kotak Oranye): Rentang interkuartil (kotak) menjadi lebih stabil. Mulai bermunculan lebih banyak outlier di bagian atas dan bawah whisker. Ini wajar karena dengan 1.000 kali pengacakan, peluang untuk mendapatkan rata-rata sampel yang sedikit menyimpang jauh dari 3.76 menjadi lebih besar.

### Bootstrap 10.000 Iterasi

Histogram (Batang/Kurva Hijau): Distribusi terlihat sangat mulus, padat, dan membentuk kurva normal yang nyaris sempurna. Sekali lagi, ini bukan berarti datanya salah, melainkan cerminan logis dari 10.000 kali penarikan sampel: semakin sering kita mengundi, semakin besar kemungkinan kita mendapatkan rata-rata yang sangat tinggi atau sangat rendah sesekali.

## Kesimpulan Singkat:
Meningkatkan jumlah iterasi dari 100 ke 10.000 tidak merubah letak nilai rata-ratanya, melainkan menyempurnakan bentuk distribusinya menjadi lebih normal dan membantu kita memetakan variasi (kemungkinan rentang kesalahan) dengan jauh lebih komprehensif.

## **4.Interval kepercayaan 95%**

```python
def calculate_ci(data):
    lower = np.percentile(data, 2.5)
    upper = np.percentile(data, 97.5)
    return lower, upper

# Menghitung CI untuk masing-masing jumlah iterasi
ci_100 = calculate_ci(boot_100)
ci_1000 = calculate_ci(boot_1000)
ci_10000 = calculate_ci(boot_10000)

# Membuat DataFrame untuk perbandingan
ci_summary = pd.DataFrame({
    'Iterasi': [100, 1000, 10000],
    'Lower Bound (2.5%)': [ci_100[0], ci_1000[0], ci_10000[0]],
    'Upper Bound (97.5%)': [ci_100[1], ci_1000[1], ci_10000[1]],
    'Rentang Interval': [ci_100[1]-ci_100[0], ci_1000[1]-ci_1000[0], ci_10000[1]-ci_10000[0]]
})

print("--- Perbandingan 95% Confidence Interval ---")
display(ci_summary)
```
